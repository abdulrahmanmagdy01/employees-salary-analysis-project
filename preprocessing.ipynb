# Cleaning Data

```python
survey = survey.drop_duplicates()
```

```python
survey.duplicated().sum()
```

```python
survey = survey.drop(columns = ['Number of vacation days','Number of home office days per month'])
```

```python
survey = survey.drop(columns = 'Company business sector')
survey
```

```python
survey["Seniority level"] = survey["Seniority level"].fillna('')
```

```python
for x in survey.index:
    if survey.loc[x, "Seniority level"] == '':
        survey.drop(x, inplace=True)
```

```python
for x in survey.index:
    if survey.loc[x, "Main technology"] == "Not Relevant":
        survey.drop(x, inplace=True)
```

```python
survey
```

```python
survey["Age"] = survey["Age"].fillna(0)
survey
```

```python
survey["Age"] = survey["Age"].apply(np.int64)
```

```python
survey["Age"] = survey["Age"].replace(0, survey["Age"].median())
```

```python
survey["Company type"] = survey["Company type"].fillna('')
```

```python
for x in survey.index:
    if survey.loc[x, "Company type"] == '':
        survey.drop(x, inplace=True)
survey
```

```python
survey["Main technology"].isna().sum()
```

```python
survey["Main technology"] = survey["Main technology"].fillna('')
```

```python
for x in survey.index:
    if survey.loc[x, "Main technology"] == '':
        survey.drop(x, inplace=True)
survey
```

```python
for x in survey.index:
    if survey.loc[x, "Main language at work"] == '':
        survey.drop(x, inplace=True)
```
